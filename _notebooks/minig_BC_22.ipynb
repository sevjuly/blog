```python
a =  pd.read_csv('data_bc/2022_01.csv', sep ='^',names=["년월","대분류","소분류","나이대","성별코드","지역","지역_2","금액지표","건수지표"])
b =  pd.read_csv('data_bc/2022_02.csv', sep ='^',names=["년월","대분류","소분류","나이대","성별코드","지역","지역_2","금액지표","건수지표"])
c =  pd.read_csv('data_bc/2022_03.csv', sep ='^',names=["년월","대분류","소분류","나이대","성별코드","지역","지역_2","금액지표","건수지표"])
d =  pd.read_csv('data_bc/2022_04.csv', sep ='^',names=["년월","대분류","소분류","나이대","성별코드","지역","지역_2","금액지표","건수지표"])
year_2022 = pd.concat([a,b,c,d])
```

# 분야별 그래프

```python
a= year_2022[year_2022['대분류']=='보건/위생']

clean= a.groupby('년월')['건수지표'].sum()
clean

clean.plot(color ="red",linewidth =3.0)
plt.ylim(2200000, 3400000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("보건/위생")
plt.show()
```

```python
b= year_2022[year_2022['대분류']=='일반음식']
b

food= b.groupby('년월')['건수지표'].sum()
food.plot(color ="orange",linewidth =3.0)
plt.ylim(3000000,7000000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("일반음식")
plt.show()
```

```python
c= year_2022[year_2022['대분류']=='유흥주점(음식)']
wine= c.groupby('년월')['건수지표'].sum()
wine.plot(color ="yellow",linewidth =3.0)
plt.ylim(0,260000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("유흥주점(음식)")
plt.show()
```

```python
d= year_2022[year_2022['대분류']=='단란주점(음식)']
wine_2= d.groupby('년월')['건수지표'].sum()
wine_2.plot(color ="yellowgreen",linewidth =3.0)
plt.ylim(0, 250000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("단란주점(음식)")
plt.show()
```

```python
e = year_2022[year_2022['대분류']=='숙박업']

house = e.groupby('년월')['건수지표'].sum()
house.plot(color ="darkgreen",linewidth =3.0)
plt.ylim(50000, 1300000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("숙박업")
plt.show()
```

```python
f= year_2022[year_2022['대분류']=='의료기관']

hospital = f.groupby('년월')['건수지표'].sum()
hospital.plot(color ="skyblue",linewidth =3.0)
plt.ylim(7000000,10500000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("의료기관")
plt.show()
```

```python
g = year_2022[year_2022['대분류']=='사무/통신기기']

tel = g.groupby('년월')['건수지표'].sum()
tel.plot(color ="darkblue",linewidth =3.0)
plt.ylim(1000000,5000000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("사무/통신기기")
plt.show()
```

```python
h = year_2022[year_2022['대분류']=='여행업']

travel = h.groupby('년월')['건수지표'].sum()
travel.plot(color ="purple",linewidth =3.0)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("여행업")
plt.show()
```

```python
i = year_2022[year_2022['대분류']=='문화/취미']

hob = i.groupby('년월')['건수지표'].sum()
hob.plot(color ="pink",linewidth =3.0)
plt.ylim(2300000,8500000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("문화/취미")
plt.show()
```

```python
j = year_2022[year_2022['대분류']=='보험']

bo = j.groupby('년월')['건수지표'].sum()
bo.plot(color ="gray",linewidth =3.0)
plt.ylim(7250000,9500000)
plt.ticklabel_format(style ='plain')
plt.xlabel("년도_월")
plt.ylabel("월별 건수 합")
plt.title("보험")
plt.show()
```

```python
comp = pd.merge(left=clean, right = food, how ="inner",on ="년월")
comp = comp.merge(right = wine, how ="inner",on ="년월")
comp = comp.merge(right = wine_2, how ="inner",on ="년월")
comp = comp.merge(right = house, how ="inner",on ="년월")
comp = comp.merge(right = hospital, how ="inner",on ="년월")
comp = comp.merge(right = tel, how ="inner",on ="년월")
comp = comp.merge(right = travel, how ="inner",on ="년월")
comp = comp.merge(right = hob, how ="inner",on ="년월")
comp
```

```python
comp.columns = ["보건/위생","일반음식","유흥주점(음식)","단란주점(음식)","숙박업","의료기관","사무/통신기기","여행업","문화/취미"]
comp
```

```python
comp_t =comp.transpose()
comp_t
```

```python
comp.plot(linewidth =3.0)  #범례 위치 변경하기
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.show()     ##1.0 은 천만 / 0.8은 8백만이다
```
